```python
import pandas as pd
import pandas as pd
import yaml
```

```python
#function to read raw data with the yaml file configuration
def read_file(yaml_path, inp_data_section, file_name):
    try:
        with open(yaml_path, "r") as file:
            cfg = yaml.safe_load(file)
    except:
        print("Yaml configuration file not found!")
        return None

    raw_df = pd.read_csv(cfg[inp_data_section][file_name])  # ✅
    return raw_df
```

```python
yaml_path= "../config.yaml"
inp_data_section = 'input_data'
file_name= 'file3'

#reading the file 3 configured in yaml file as file3: "../data/raw/df_final_experiment_clients.txt"
file3= read_file(yaml_path,inp_data_section,file_name)
f3=file3.copy()
f3.head()
```

```python
# Merge the sales and revenue DataFrames on the 'Date' column (inner join)
# Only rows with a common value in the 'client_id' column, present in both DataFrames, are included in the merged result.
df=pd.merge(fexp, f2, on='client_id')
display(df.info())
display(df.head(2))
```

```python
def out_csv(df,yaml_path,output_section_yaml,file_name):
    try:
        with open(yaml_path, "r") as file:
            cfg = yaml.safe_load(file)
    except:
        print("Yaml configuration file not found!")
        return None
        
    df.to_csv(cfg[output_section_yaml][file_name], index=False)
    print(f"File saved to: {cfg[output_section_yaml][file_name]}")
```

```python
#exporting csv to the clean data folder as specified with the yaml file:   file2: "../data/clean/digfootpring_groups_clean.csv"

yaml_path= "../config.yaml"
output_section_yaml='output_data'
out_file_name= 'file2'    

out_csv(df,yaml_path,output_section_yaml,out_file_name)
```
