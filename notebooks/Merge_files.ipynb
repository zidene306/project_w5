```python
#Mergin datasets
import pandas as pd
import pandas as pd
import yaml

#function to read raw data with the yaml file configuration
def read_file(yaml_path, inp_data_section, file_name):
    try:
        with open(yaml_path, "r") as file:
            cfg = yaml.safe_load(file)
    except:
        print("Yaml configuration file not found!")
        return None

    raw_df = pd.read_csv(cfg[inp_data_section][file_name])  # ✅
    return raw_df
```

```python
yaml_path= "../config.yaml"
inp_data_section = 'output_data'
f1= 'file1' #clean_final_demo.csv
f2='file2' #digfootpring_groups.csv
```

```python
#importing the two files

f1= read_file(yaml_path,inp_data_section,f1)
f2= read_file(yaml_path,inp_data_section,f2)

display(f1.head(1))
display(f2.head(1))
```

```python
f1.info()
print('________')
f2.info()
```

```python
merged_df = pd.merge(f1, f2, on='client_id', how='outer')
display(merged_df.head(2))
print('__________')
merged_df.info()
```
