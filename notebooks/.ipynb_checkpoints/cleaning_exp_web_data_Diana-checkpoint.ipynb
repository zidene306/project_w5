```python
import pandas as pd

#We import the data set that tells who was part of the experiment, as those will be the clients to investigate.
file3 = pd.read_csv('df_final_experiment_clients.txt')
f3=file3.copy()
f3.head()
```

```python
display(f3.info())
print('____')
display(f3.isnull().sum())
f3.duplicated().sum()
```

```python
#data frame containint only those client_ids which were part of the experiment, either in Test group or Control group
#fexp=f3.dropna() 
fexp=f3
fexp.info()
```

```python
#importing the two files
file2p1 = pd.read_csv('df_final_web_data_pt_1.txt')
file2p2 = pd.read_csv('df_final_web_data_pt_2.txt')

f2p1=file2p1.copy() #file part 1
f2p2=file2p2.copy() #file part 2
```

```python
f2p1.info()
print('____________')
f2p2.info()
```

```python
display(f2p1.head(1))
display(f2p2.head(1))
```

```python
#We check if there are duplicated rows
display(f2p1.duplicated().sum())
f2p1[f2p1.duplicated(keep=False)].head(4)  # shows ALL copies, not just 2nd occurrence
```

```python
f2p1=f2p1.drop_duplicates()
f2p1.shape
```

```python
#We check if there are duplicated rows for the second data set
display(f2p2.duplicated().sum())
f2p2[f2p2.duplicated(keep=False)].head(4)  # shows ALL copies, not just 2nd occurrence
```

```python
f2p2=f2p2.drop_duplicates()
f2p2.shape
```

```python
# Concatenating files vertically (along rows)
f2=pd.concat([f2p1, f2p2], axis=0).reset_index(drop=True)
display(f2.info())
display(f2.head(2))
```

```python
# Merge the sales and revenue DataFrames on the 'Date' column (inner join)
# Only rows with a common value in the 'client_id' column, present in both DataFrames, are included in the merged result.
df=pd.merge(fexp, f2, on='client_id')
display(df.info())
display(df.head(2))
```
