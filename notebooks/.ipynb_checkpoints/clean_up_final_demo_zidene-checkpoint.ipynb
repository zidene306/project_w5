```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import yaml
#import functions
output_dir ="../data/clean"
with open("../config.yaml") as file:
    config = yaml.safe_load(file)
config
```

```python
df_final_demo = pd.read_csv(config["input_data"]["file1"])
df_final_web_data = pd.read_csv(config["output_data"]["file2"])
#df_final_web_data.head()
#df_final_demo .info()
#df_final_demo .shape
```

```python
df_final_web_data.info()
```

```python
def clean_final_demo_func(df_final_demo):
    """
    cleans df_final_demo.txt and outputs df_final_demo_cleaned.csv
    """
    df = df_final_demo.copy()
    #drop empty rows
    df = df.dropna(subset = "clnt_tenure_yr", ignore_index = True)
    #replace the missing age value in  clnt_age F
    condition = df["gendr"] == "F"
    #female_df = df_final_demo[condition]
    mean_age_f = df[condition].clnt_age.mean()
    
    df['clnt_age'] = df.clnt_age.fillna(mean_age_f)
    
    return df
clean_final_demo= clean_final_demo_func(df_final_demo)
#clean_final_demo.to_csv("clean_final_demo.csv", index = False)
```

```python
##############################################################################
#MERGE DF FUNCTION: cast date_time to date 
##############################################################################
def merge_dfs_funct(df_final_demo, df_final_web_data):
    
    df = df_final_demo.merge(df_final_web_data, on = "client_id", how = "outer")

    #clean date columns
    df.date_time = pd.to_datetime(df.date_time, format = '%Y-%m-%d %H:%M:%S', errors = 'coerce')
    #fillup Variation column empty values with: no-data
    df["Variation"] = df.Variation.fillna("no_data")
    return df

df = merge_dfs_funct(df_final_demo, df_final_web_data)

df.to_csv(f"{output_dir}/cleaned_dataser.csv", index=False, encoding= "utf-8", sep = ";")
#Who are the primary clients using this online process?
#df.head()
#df[df.client_id == 1028]
df.head()
```

```python
##############################################################################
# PLOT1: users with complete steps
##############################################################################


user_step_counts = df.groupby('client_id')['process_step'].nunique()

step_distribution = user_step_counts.value_counts().sort_values(ascending = False)
step_distri_perc = user_step_counts.value_counts(normalize = True).sort_values(ascending = False)

fig, axes = plt.subplots(figsize=(8,5))
ax = sns.barplot(x = step_distribution.index,
            y = step_distribution.values,
            hue = step_distribution.index,
            legend = False,
            palette = ("Set3")
            )
# Add labels on bars
for container in ax.containers:
    ax.bar_label(container)
    
plt.title("Distribution of Process Steps Completed")
plt.xlabel("Number of Unique Steps Completed")
plt.ylabel("Number of Users")
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
#show numbers

plt.show()
##############################################################################
# PLOT2 proportion of completed users
##############################################################################
step_distri_perc = user_step_counts.value_counts(normalize = True).sort_values(ascending = False)
pie = step_distri_perc.plot.pie(autopct = '%1.1f%%', startangle=90, colors=sns.color_palette("Set3"))
plt.show()
step_distri_perc
```
